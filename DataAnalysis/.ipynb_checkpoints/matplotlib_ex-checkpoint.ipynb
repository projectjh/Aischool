## subplot 명령


```python
x1 = np.linspace(0.0, 5.0, num=50)
x2 = np.linspace(0.0, 2.0, num=50)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.suptitle('*** Main Title ***')

plt.subplot(2, 1, 1)
# plt.subplot(211)
plt.plot(x1, y1, 'yo-')
plt.title('A table of 2 subplots')
# 여기서 title은 figure에 대한 타이틀이 아닌 해당 axes에 대한 title임
plt.xlabel('time (s)')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.title('B table of 2 subplots')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()

# 전체 figure에 대한 타이틀을 부여하고 싶을때에는 suptitle()이라는 명령어를 사용해준다.
```

### 컨투어 플롯(contour plot)

#### np.meshgrid()

meshgrid 명령은 사각형 영역을 구성하는 가로축의 점들과 세로축의 점을 나타내는 두 벡터를 인수로 받아서 이 사각형 영역을 이루는 조합을 출력한다. 결과는 그리드 포인트의 x값 만을 표시하는 행렬과 y값을 표시하는 행렬 두 개로 분리하여 출력한다.

```python
x = np.arange(3)
y = np.arange(5)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
X
# X는 x가 가지고 있는 데이터를 행 방향으로 반복
```

```python
Y
# Y는 y가 가지고 있는 데이터를 열 방향으로 반복
```

```python
def f(x, y):
    return(1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=0.75, cmap='rainbow')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()

# np.linspace(-3, 3, n)
# -3부터 3까지 일정한 간격으로 256개를 만들겠다는 의미
# alpha => 투명도
# cmap 의 색상은 matplotlib 홈페이지에서 볼 수 있다.
```

### 바 차트(bar chart)

```python
irum = ['hong', 'kim', 'lee']
jumsu = [90, 70, 80]
n_data = len(irum)
index = np.arange(n_data)
c = ['r', 'g', 'b']
plt.bar(index, jumsu, color=c, tick_label=irum)
plt.show()
```

```python
font3 = {'family': font_name, 'color': '#333', 'size': 13}

y1 = [70, 90, 80]
y2 = [60, 70, 90]

x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5
err2 = np.abs(np.random.randn(3)) + 5

xlabel = ['국어', '영어', '수학']
plt.title("Bar Chart")
plt.bar(x, y1, yerr=err1, ecolor='yellow', width=0.3, color='red', align='center', label='점수')
# x 위치, y1 위치, err: error => yerr는 y축의 error 라고 보면 되며, 이는 표준편차로 보면 된다.
# 현재 err1은 시각화를 위한 예시일뿐 실제 데이터는 표준편차를 계산하여 넣어주면 된다.
plt.bar(x+0.4, y2, yerr=err2, width=0.3, color='blue', align='center', label='반평균')
# x+0.4는 막대그래프들의 간격을 나타내기 위해 사용한 값
plt.xticks(x+0.4/2, xlabel)
# xtics는 가운데를 찾기 위한 계산 방법
plt.ylim(0, 100)
plt.xlabel("과목", fontdict=font3)
plt.ylabel("점수", fontdict=font3)
plt.legend()
plt.show()
```

```python
y1 = [70, 90, 80]
y2 = [60, 70, 90]

x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5
err2 = np.abs(np.random.randn(3)) + 5

xlabel = ['국어', '영어', '수학']
plt.title("Bar Chart")
plt.barh(x, y1, xerr=err1, ecolor='yellow', height=0.3, color='red', align='center', label='점수')
plt.barh(x+0.3, y2, xerr=err2, height=0.3, color='blue', align='center', label='반평균')
plt.yticks(x+0.3/2, xlabel)
plt.xlim(0, 100)
plt.xlabel("점수", fontdict=font3)
plt.ylabel("과목", fontdict=font3)
plt.legend()
plt.show()
```

### 파이 차트 (pie chart)

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15.55, 30, 45, 9.45]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', shadow=True, startangle=90, counterclock=False)
plt.axis('equal')
plt.show()

# autopct='%1.2f%%'
# %1.2f 소수점 두자리 실수
# 뒤의 %%는 %문자를 표기하기 위해 사용

# counterclock=False (시계방향)
# default값 => counterclock=True (시계반대방향)
```

```python
import seaborn as sns
```

```python
# titanic = sns.load_dataset("titanic")   # 타이타닉 호 데이터
# tips = sns.load_dataset("tips")         # 팁 데이터
# flights = sns.load_dataset("flights")   # 여객운송 데이터

iris = sns.load_dataset("iris")           # 붓꽃 데이터 (분류용)
iris

# 일반적인 데이터를 회귀용, 분류용으로 구분할 수 있다.

# 붓꽃 데이터는 앞의 4가지 데이터를 가지고 타깃(종을 구분)을 구하는 분류용 데이터이다. 
# 꽃받침(sepal) 길이, 꽃받침 너비, 꽃잎(petal) 길이, 꽃잎 너비
# 품종은 3가지 존재 - setosa / versicolor / virginica
# 현재 파일에 품종별 50개씩 데이터가 존재한다.
```

#### rugplot 

```python
x = iris.petal_length.values
sns.rugplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Rug Plot")
plt.show()
```

#### kdeplot

```python
sns.kdeplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Kernel Density Plot")
plt.show()
```

### distplot

```python
sns.distplot(x, kde=True, rug=True)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Dist Plot")
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="hex", color='y')
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02)
plt.show()
```

### heatmap

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
titanic_size = titanic.pivot_table(index="class", columns="sex", aggfunc="size")
sns.heatmap(titanic_size, cmap=sns.light_palette("gray", as_cmap=True), annot=True, fmt="d")
plt.title("Heatmap")
plt.show()

# annot=True : 우측의 색상값을 표현하겠다는 의미
# fmt: 포맷형식
# fmt='d' : 값의 형태를 숫자형태로 표기하겠다는 의미 (박스 안의 숫자)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 전체 팁의 Box Plot")
plt.show()

# 박스 내부의 가로선은 중앙값
# 박스 위의 선은 Q3(제 3사분위수)
# 박스 아래 선은 Q1(제 1사분위수)
# 박스 바깥 가장 위 선 = Q3 + (IQR * 1.5)
# 박스 바깥 가장 아래 선 = Q1 - (IQR * 1.5)
```

### violinplot

```python
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 전체 팁의 Violin Plot")
plt.show()
```

### stripplot

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("요일 별 전체 팁의 Strip Plot")
plt.show()

# jitter=True 겹치지 않도록 무작위로 이동시켜줌
```

### swarmplot

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 전체 팁의 Swarm Plot")
plt.show()
```

### catplot

```python
data = titanic[titanic.survived.notnull()]
sns.catplot(x="age", y="sex", hue="survived", row="class", data=data, kind="violin", split=True, height=2, aspect=4)
plt.title("각 클래스의 성별 생존자 수의 Catplot")
plt.show()
```

### 여러 종류 차트 겹쳐서 표시

```python
plt.title("Boxplot과 Strip Plot으로 표현한 요일별 팁")
sns.boxplot(x="tip", y="day", data=tips, whis=np.inf)
sns.stripplot(x="tip", y="day", data=tips, jitter=True, color="0.4")
plt.show()
```
