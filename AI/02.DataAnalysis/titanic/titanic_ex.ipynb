```python
titanic_df['Age_by_10'] = titanic_df['Age'] * 10
titanic_df['Family_No'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
titanic_df.head(3)
# Age_by_10, Family_No 컬럼이 없으므로 컬럼 생성후 계산된 값 저장
```

```python
titanic_df['Age_by_10'] = titanic_df['Age_by_10'] + 100
titanic_df.head(3)
# 컬럼이 존재하므로 계산 값 수정
```

```python
titanic_drop_df = titanic_df.drop('Age_0', axis=1)
titanic_drop_df.head(3)
# axis=1 열방향(컬럼)
# 2차원일때는 0, 1로 행, 열을 표현할 수 있는데,
# 3차원에서는 0, 1을 사용하면 달라짐
# 행이 큰차원, 열이 작은차원
# 2차원 구조에서 가장 큰 차수는 행
# 3차원 구조는 면, 행, 열로 표현되며, axis=0 면 / axis=1 행 / axis=2 열 을 가리킨다.
# 음수로 표현하는 경우 -1이 가장 낮은 차수부터 시작이 된다
# 그러므로 
# 2차원에서는 axis=-1 열 / axis=-2 행
# 3차원에서는 axis=-1 열 / axis=-2 행 / axis=-3 면
```

```python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
# fillna() => fill+na
# NaN의 결측치에 해당 데이터로 값을 채워라는 의미 
# 평균값, 중앙값 등의 내용으로도 채울 수 있다.
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')
titanic_df.isna().sum()
```

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(3)
# map()은 하나의 컬럼에 대한 연산,
# apply()는 하나 이상의 컬럼과 하나 이상의 파라미터가 있을때 사용
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: 'Child' if x <=15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
titanic_df[['Age_cat']]
```

```python
def get_category(age):
    cat = ''
    if age <= 5:
        cat = 'Baby'
    elif age <= 12:
        cat = 'Child'
    elif age <= 18:
        cat = 'Teenager'
    elif age <= 25:
        cat = 'Student'
    elif age <= 35:
        cat = 'Young Adult'
    elif age <= 60:
        cat = 'Adult'
    else:
        cat = 'Elderly'
        
    return cat

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
