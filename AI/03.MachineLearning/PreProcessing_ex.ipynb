## 01_표준화

```python
import seaborn as sns

iris = sns.load_dataset("iris")
iris

# seaborn은 데이터를 데이터프레임형식으로 제공
```

```python
x = iris.iloc[:, :-1]
x

# 행은 전체, 열은 species를 제외한 나머지
# 이 자료를 가지고 표준화 작업진행
# 표준화하는 것이 바람직한지 아닌지 결과값을 보고 판단
```

#### scale()

```python
from sklearn.preprocessing import scale

x_scaled = scale(x)
x_scaled[:5, :]

# 표준화 작업을 거치고 나면 numpy배열이 반환된다.
```

```python
x_scaled.mean(axis=0)
```

```python
for scaled_mean in x_scaled.mean(axis=0):
    print("{:10.9f}".format(scaled_mean))
    
# 소수점아래 9자리 수가 모두 0으로 나오기때문에 사실상 0에 가까운 수라고 할 수 있다.
```

#### robust_scale()

```python
from sklearn.preprocessing import robust_scale

iris_robust_scaled = robust_scale(x)
iris_robust_scaled[:5, :]
```

#### minmax_scale()

```python
from sklearn.preprocessing import minmax_scale

iris_minmax_scaled = minmax_scale(x)
iris_minmax_scaled[:5, :]
```

#### maxabs_scale()

```python
from sklearn.preprocessing import maxabs_scale

iris_maxabs_scaled = maxabs_scale(x)
iris_maxabs_scaled[:5, :]
```

### StandardScaler

```python
iris = sns.load_dataset("iris")
iris.iloc[:5, :-1]
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
iris_scaled = sc.fit_transform(iris.iloc[:, :-1])
iris_scaled[:5, :]
```

```python
iris_origin = sc.inverse_transform(iris_scaled)
iris_origin[:5, :]
```

### MinMaxScaler

```python
iris_origin = sc.inverse_transform(iris_scaled)
iris_origin[:5, :]
```

### 원-핫 인코딩

#### * 원-핫 인코딩으로 변환시킬 수 있는 3가지 방법

- OneHotEncoder: 사이킷런
- get_dummies(): 판다스
- to_categorical(): 케라스



Q. 레이블인코딩이 용량도 적게 차지하고 보기에도 쉬운데 왜 원-핫 인코딩을 사용할까?<br>
A. 레이블인코딩은 숫자로 구성되어있기때문에 y = wx + b 수식을 계산할때 x값에 들어가서 영향을 미칠 수 있기때문에 원-핫 인코딩을 사용하는 것이 좋다.<br>
   <span style="color:#ff6600; font-weight:600; padding-left:20px">∴ 클래스가 많은 데이터를 사용할때는 원-핫 인코딩을 사용하는 것이 좋다.</span>
   


```python
iris[['species']]
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
enc.fit(iris[['species']])

# OneHotEncoder사용시 2차원 구조로 만들어줘야함
```

```python
iris_onehot = enc.transform(iris[['species']])
print(iris_onehot)
iris_onehot

# sparse matrix 희소행렬: 0이 많은 행렬
# 이를 numpy 배열로 바꾸면 확인할 수 있음

# print(iris_onehot)
# 희소행렬은 해당 값의 좌표와 값을 저장한다.
# ex) 0행 0열 값 1.0
#     1행 0열 값 1.0
# 학습에 사용하기 위해 아래와 같이 toarray()를 사용하여 변환시켜야 한다.
```

```python
iris_onehot.toarray()
```

### 결측값 처리

```python
import seaborn as sns

iris = sns.load_dataset("iris")
x = iris.iloc[:, :-1]
x
```

```python
import random

for col in range(4):
    x.iloc[[random.sample(range(len(iris)), 10)], col] = float('nan')
    # 매번 컬럼별로 무작위로 10개씩 nan을 대입
    
x.head()
```

```python
x.mean(axis=0)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit_transform(x)[0:5, :]

# fit_transform() 평균을 계산해서 값을 대체
# strategy 옵션값
#   mean: 평균값
#   median: 중간값
#   most_frequent: 최빈값
```

### Q. 타이타닉 데이터에 Age, Embarked 결측값 처리 해보기

Age => 평균 or 중앙값<br>
Embarked => 최빈값

```python
df = pd.read_csv("../2_DataAnalysis/titanic/train.csv")
df.head()
```

```python
df_age = df[['Age']]
df_age
```

```python
imp_age = SimpleImputer(missing_values=np.nan, strategy='median')
df[['Age']] = imp_age.fit_transform(df_age)
df
```

```python
df_embarked = df[['Embarked']]
df_embarked
```

```python
imp_embarked = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df[['Embarked']] = imp_embarked.fit_transform(df_embarked)
# df
imp_embarked.fit_transform(df_embarked)
```
