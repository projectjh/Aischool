### 케라스(keras)를 이용한 선형회귀 예제

#### 1. 라이브러리 임포트 및 학습데이터 정의

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD
import numpy as np

x_data = np.array([1, 2, 3, 4, 5, 6])
t_data = np.array([3, 4, 5, 6, 7, 8])
print(tf.__version__)
```

#### 2. 모델(model) 구축

```python
model = Sequential()
model.add(Flatten(input_shape=(1,)))   # 입력층 (1차원 입력)
model.add(Dense(1, activation='linear'))   # 출력층

# model.add(Dense(1, input_shape=(1,), activation='linear'))   # 입력층 + 출력층

# Flatten(): 모든 데이터를 1차원 구조로 만들어줌
# ✨ actication => 활성함수 (선형(회귀)모델 => linear / 분류모델 => 이진분류: sigmoid, 다중분류: softmax)
```

#### 3. 모델 컴파일 및 요약

```python
model.compile(optimizer=SGD(learning_rate=1e-2), loss="mse")
model.summary()

# SGD: 획률적 경사하강법 => bias와 weight를 업데이트할 수 있는 최적화 알고리즘
# 회귀모델이므로 손실값을 구할때 mse사용
# 입력파라미터 1개, 출력파라미터 1개로 dense의 Param은 총 2개가 된다.
```

#### 4. 모델 학습

```python
hist = model.fit(x_data, t_data, epochs=1000)

# 모든 데이터를 가지고 한 번 입->출력 과정을 거치는 것을 1epoch라고 한다.
```

#### cf) 차트그려서 확인

```python
hist.history
```

```python
import matplotlib.pyplot as plt

plt.title("Loss Trend")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.grid()

plt.plot(hist.history['loss'], label="train loss")
plt.legend(loc="best")

plt.show()
```

#### 5. 모델 평가 및 예측

```python
result = model.predict(np.array([-3.1, 3.0, 3.5, 15.0, 20.1]))
print(result)
```
