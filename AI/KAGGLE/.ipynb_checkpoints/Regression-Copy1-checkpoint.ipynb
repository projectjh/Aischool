```python
import numpy as np

X_train = np.array([1, 2, 3, 4, 5]).reshape(5,1) # 입력데이터 초기화
X_test = np.array([1, 2, 3, 4, 5]).reshape(5,1)
y_train = np.array([2, 3, 4, 5, 6]).reshape(5,1) # 정답데이터 초기화
W = np.random.rand(1,8) # 가중치 W 초기화
b = np.random.rand(1) # 바이어스 b 초기화

def loss_func(x, t):# 손실함수
    y = np.dot(x,W) + b
    return (np.sum((t - y)**2 )) / (len(x))

# 수치미분 함수
def numerical_derivative(f, x):
    delta_x = 1e-4 # 0.0001
    grad = np.zeros_like(x) 
    it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])

    while not it.finished:
        idx = it.multi_index 
        tmp_val = x[idx]
        x[idx] = float(tmp_val) + delta_x
        fx1 = f(x) # f(x+delta_x)

        x[idx] = float(tmp_val) - delta_x 
        fx2 = f(x) # f(x-delta_x)
        grad[idx] = (fx1 - fx2) / (2*delta_x)
 
        x[idx] = tmp_val 
        it.iternext()
    return grad

learning_rate = 1e-2 # 학습율 α
f = lambda x : loss_func(X_train, y_train) 
print("Initial loss value = ", loss_func(X_train, y_train), "Initial W = ", W, "\n", ", b = ", b )

for step in range(10001):
    W -= learning_rate * numerical_derivative(f, W) 
    b -= learning_rate * numerical_derivative(f, b)
    if (step % 200 == 0): # 추가부분
        print("step = ", step, "loss value = ", loss_func(X_train, y_train), "W =", W, "b =", b)
```

```python
def predict(x): # 예측 함수
    y = np.dot(x,W) + b
    return y
predict(X_test) # 예축 테스트
```
