```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action="ignore")
```

```python
df_train = pd.read_csv('./data/train.csv')
df_test = pd.read_csv('./data/test.csv')
```

```python
df_train
```

```python
df_test
```

```python
df_train.loc[(df_train['GARAGE'] >= 5.0), ['GARAGE']] = 5.0
```

```python
df_train['GARAGE'].value_counts()
```

```python
df_train.isnull().sum()
```

```python
df_train['GARAGE'].mode()
```

```python
df_train['GARAGE'] = df_train['GARAGE'].fillna(df_train['GARAGE'].mode()[0])
df_train.isnull().sum()
```

```python
df_train.drop(['ID', 'ADDRESS', 'SUBURB', 'LAND_AREA', 'POSTCODE', 'LATITUDE', 'LONGITUDE', 'NEAREST_STN', 'NEAREST_SCH'], axis=1, inplace=True)
df_train
```

```python
df_train.isnull().sum()
```

```python
df_train.drop(['NEAREST_SCH_RANK', 'BUILD_YEAR'], axis=1, inplace=True)
df_train
```

```python
df_train2 = df_train.dropna(axis=0)
df_train2.isnull().sum()
```

```python
df_train2.reset_index(drop=True, inplace=True)
df_train2
```

```python
df_train2.info()
```

```python
df_train2['DATE_SOLD'] = pd.to_datetime(df_train2.DATE_SOLD)
df_train2
```

```python
month = pd.to_datetime(df_train2['DATE_SOLD'], format='%m')
month
```

```python
df_train2['DATE_SOLD'] = df_train2['DATE_SOLD'].dt.month
df_train2
```

```python
df_train3 = df_train2.drop(['DATE_SOLD'], axis=1)
df_train3
```

---

#### MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()
df_train2.iloc[:, 1:] = sc.fit_transform(df_train2.iloc[:, 1:])
df_train2
```

#### standardScaler

```python
from sklearn.preprocessing import StandardScaler

ssc = StandardScaler()
df_train2.iloc[:, 1:] = ssc.fit_transform(df_train2.iloc[:, 1:])
df_train2
```

#### MaxAbsScaler

```python
from sklearn.preprocessing import MaxAbsScaler

mas = MaxAbsScaler()
df_train3.iloc[:, 1:] = mas.fit_transform(df_train3.iloc[:, 1:])
df_train3
```

---

#### data검증

```python
x_data = df_train3.iloc[:, 1:]
y_data = df_train3['PRICE']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, random_state=0)
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
