```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action="ignore")

df_train = pd.read_csv("./data/classification/train.csv")
df_test = pd.read_csv("./data/classification/test.csv")
```

```python
train1 = df_train.drop(['no', 'education', 'native-country'], axis=1)
train1
```

```python
train1.isnull().sum()
```

```python
train1['workclass'].values
```

```python
train1[train1['workclass'] == " ?"]
```

```python
train1[(train1['workclass'] == " ?") & (train1['income'] == 0)]
```

```python
train1[(train1['workclass'] == " ?") & (train1['income'] == 1)]
```

```python
train1[(train1['workclass'] == " ?") | (train1['occupation'] == " ?")]
```

```python
train1.loc[(train1['workclass'] == ""), ['workclass']] = None
train1
```

```python
train1 = train1.dropna(axis=0)
train1
```

```python
train1.reset_index(drop=True, inplace=True)
train1.tail()
```

```python
train1.corr()
```

```python
train1['sex'].values
```

```python
train1['sex'].replace({' Male': 0, ' Female': 1}, inplace=True)
train1.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
train1['workclass'] = le.fit_transform(train1['workclass'])
train1
```

```python
train1['marital-status'] = le.fit_transform(train1['marital-status'])
train1['occupation'] = le.fit_transform(train1['occupation'])
train1['relationship'] = le.fit_transform(train1['relationship'])
train1['race'] = le.fit_transform(train1['race'])

train1
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
train1[['fnlwgt']] = mms.fit_transform(train1[['fnlwgt']])
train1.iloc[:, -4:-2] = mms.fit_transform(train1.iloc[:, -4:-2])
train1.head()
```

```python
x_data = train1.iloc[:, :-1]
y_data = train1['income']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, random_state=11)
X_train.shape, X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=10) # max_depth : 트리의 최대 깊이
model.fit(X_train, y_train)
```

```python
score = model.score(X_test, y_test)
print("정확도 :", score)
```
