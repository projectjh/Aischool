```python
from patsy import dmatrices

y, X = dmatrices(formula, df_train, return_type="dataframe")
y
```

```python
def get_vif(formula, df):
    from patsy import dmatrices
    y, X = dmatrices(formula, df, return_type="dataframe")
    vif = pd.DataFrame()
    
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    
    vif['VIF Factor'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['features'] = X.columns
    
    return vif

formula = "Survived~Pclass+Sex+Age+SibSp+Parch+Fare+Embarked-1"
get_vif(formula, df_train)
```

```python
import statsmodels.formula.api as smf

model_titanic = smf.ols(formula=formula, data=df_train).fit()
model_titanic.summary()
```

## Answer

```python
df = pd.read_csv("../2_DataAnalysis/titanic/train.csv")
df
```

##### Passengerid, Name, Ticket, Cabin 컬럼 삭제

```python
df2 = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
df2
```

##### NaN 데이터 삭제✨

```python
df2 = df2.dropna(axis=0).copy()
df2

# dropna() 축방향에다라 NaN 데이터 삭제
# copy()는 복사를 위해 사용 => inplace=True를 사용해도 됨
```

##### 기존 인덱스 재정렬

```python
df2.reset_index(drop=True, inplace=True)
df2

# 기존 인덱스를 삭제 후 재정비
```

##### Sex, Embarked는 레이블 인코딩

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df2_Sex = le.fit_transform(df2['Sex'])
df2['Sex'] = df2_Sex
df2.head()
```

```python
le = LabelEncoder()
df2_Embarked = le.fit_transform(df2['Embarked'])
df2['Embarked'] = df2_Embarked
df2.head()
```

##### Age, Fare는 MinMaxScaler로 전처리

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()
mm_Age = mm.fit_transform(df2[['Age']])
df2['Age'] = mm_Age

mm_Fare = mm.fit_transform(df2[['Fare']])
df2['Fare'] = mm_Fare
df2.head()
```

```python
import statsmodels.formula.api as smf

model_titanic = smf.ols(formula="Survived~Pclass+Sex+Age+SibSp+Parch+Fare+Embarked", data=df2).fit()
model_titanic.summary()
```

```python
def get_vif(formula, df):
    from patsy import dmatrices
    y, X = dmatrices(formula, df, return_type="dataframe")
    vif = pd.DataFrame()
    
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    
    vif['VIF Factor'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['features'] = X.columns
    
    return vif

formula = "Survived~Pclass+Sex+Age+SibSp+Parch+Fare+Embarked"
get_vif(formula, df2)
```
